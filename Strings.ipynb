```python
3) String Slicing(Extra Important)
Slicing means cutting part of string
Syntax:
string[start:end]
start=included
end=not included
```

```python
name = "PYTHON"

print(name[0:3])
print(name[2:5])
print(name[:4])
print(name[3:])
```

```python
#Reverse a string
name = "PYTHON"
print(name[::-1])
```

```python
#1.Count Characters(Without len())
#Question
#Count the number of chrcters in string without using len().
text="Python"
count=0

for ch in text:
    count +=1

print("Total charcters:",count)
```

```python
#2.Count Words in a Sentence
#Question
#Count the total number of words in a sentence.
sentence="Python is very easy"
words=sentence.split()
print("Total words:",len(words))
```

```python
#3.Count Uppercase and Lowercase letters
#Question
#Count uppercase and lowercase letters separately.
text = "PyThOn"
upper = 0
lower = 0

for ch in text:
    if ch.isupper():
        upper += 1
    elif ch.islower():
        lower += 1

print("Uppercase:", upper)
print("Lowercase:", lower)
```

```python
#4.Count Digits,Letters,Special Characters
#Question
#Count digits,alphabets,and special characters in string.
text="Python123@"
letters=digits=special=0

for ch in text:
    if ch.isalpha():
        letters += 1
    elif ch.isdigit():
        digits += 1
    else:
        special += 1

    print("Letters:",letters)

    print("Digits:",digits)
    print("Special charcters:",special)
    
```

```python
#5.Remove Duplicate characters
#Question
#Remove duplicate charcters from a string.
text="programming"
result=""

for ch in text:
    if ch not in result:
        result+=ch

        print("result:",result)
```

```python
#6.Check Palindrome(Without Slicing)
#Question
#Check whether a word is a palindrome without using slicing.
word="madam"
rev=""

for ch in word:
    rev=ch+rev

if word==rev:
    print("Palindrome")
else:
    print("Not Palindrome")
```

```python
#7.Frequency of Each Charater
#Question
#Print frequency of each character in string.
text="hello"

for ch in set(text):

    print(ch,":",text.count(ch))
```

```python
#8.Check Anagram
#Question
#Check if two strings are anagrams.
s1 = "listen"
s2 = "silent"

if sorted(s1) == sorted(s2):
    print("Anagram")
else:
    print("Not Anagram")
```

```python
#9.Find Longest world in a sentence
#Question
#Find the longest word in a sentence.
sentence="Python programming is powerful"
words=sentence.split()

longest=words[0]

for word in words:
    if len(word)>len(longest):
        longest=word

    print("Longest word:",longest)
```

```python
#10.Password Validation
#Question
#check if password is valid:
#minimum 8 chracters
#at least 1 digit
#at least 1 uppercase letter
password = "Python123"
digit = upper = False

for ch in password:
    if ch.isdigit():
        digit = True
    if ch.isupper():
        upper = True

if len(password) >= 8 and digit and upper:
    print("Valid Password")
else:
    print("Invalid Password")
```
